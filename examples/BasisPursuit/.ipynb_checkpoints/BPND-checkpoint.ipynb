## Sparco Problem 2: blocksig

```julia
m = 1024
n = 1024
lines = readlines("./prob2.txt", keep=true)
b = Vector{Float64}()
for line in lines
    push!(b, parse(Float64, line))
end
# α = (norm(b)*1e-3)^2/2
M = LinearMap(x->idwt(x, wavelet(WT.haar), 5), y->dwt(y, wavelet(WT.haar), 5), 1024, 1024)
k = 71
τ = 4.5e+2
A = OneBall(n, k)
```

```julia
dual_gaps = Vector{Float64}()
pr_infeas = Vector{Float64}()
sol = level_set(M, b, A; 
    α=0.0, tol=1e-3, maxIts=100, τmax=τ, dual_gaps=dual_gaps, infeas=pr_infeas); 
```

```julia
x = constructPrimal(sol)
@printf "Relative recovery error: %.4f" norm(M*x - b) / norm(b)
```

## Sparco Problem 3: cosspike

```julia
m = 1024
n = 2048
lines = readlines("./prob3.txt", keep=true)
b = Vector{Float64}()
for line in lines
    push!(b, parse(Float64, line))
end
α = (norm(b)*1e-3)^2/2
M1 = LinearMap(x->idct(x), y->dct(y), 1024, 1024)
M2 = I(1024)
M = hcat(M1, M2)
k = 113
τ = 2.2e+2
A = OneBall(n, k)
```

```julia
dual_gaps = Vector{Float64}()
pr_infeas = Vector{Float64}()
sol = level_set(M, b, A; 
    α=0.0, tol=1e-3, maxIts=20000, τmax=τ, dual_gaps=dual_gaps, infeas=pr_infeas);
```

```julia
x = constructPrimal(sol)
@printf "Relative recovery error: %.4f" norm(M*x - b) / norm(b)
```

```julia
gr(size=(500,400))
rounds = collect(1:length(dual_gaps))
p = plot(rounds, dual_gaps, label="dual gap", yaxis=:log, line=2)
plot!(rounds, pr_infeas, label="recovery error", yaxis=:log, line=2)
```

## Sparco problem 903: spiketrn

```julia
m = 1024
n = 1024
lines = readlines("./prob903_obs.txt", keep=true)
b = Vector{Float64}()
for line in lines
    push!(b, parse(Float64, line))
end
# α = (norm(b)*1e-3)^2/2
lines = readlines("./prob903_kernel.txt", keep=true)
kernel = Vector{Float64}()
for line in lines
    push!(kernel, parse(Float64, line))
end
l = length(kernel)
function f1(x)
    xf = fft(vcat(x, zeros(l)))
    kf = fft(vcat(kernel, zeros(n)))
    out = ifft( xf .* kf)
    return real(out[1:n])
end

function f2(x)
    xf = fft(vcat(x, zeros(l)))
    kf = conj(fft(vcat(kernel, zeros(n))))
    out = ifft( xf .* kf)
    return real(out[1:n])
end

M = LinearMap(x->f1(x), y->f2(y), n, n)
τ = 1.3e+1
k = 35
A = OneBall(n, k)
```

```julia
dual_gaps = Vector{Float64}()
pr_infeas = Vector{Float64}()
sol = level_set(M, b, A; 
    α=0.0, tol=1e-3, maxIts=20000, τmax=τ, dual_gaps=dual_gaps, infeas=pr_infeas);
```

```julia
x = constructPrimal(sol)
@printf "Relative recovery error: %.4f" norm(M*x - b) / norm(b)
```

```julia
gr(size=(500,400))
rounds = collect(1:length(dual_gaps))
p = plot(rounds, dual_gaps, label="dual gap", yaxis=:log, line=2)
plot!(rounds, pr_infeas, label="recovery error", yaxis=:log, line=2)
```
