```julia
using AtomicOpt
using LinearAlgebra
using SparseArrays
using Printf
using Arpack
using LinearMaps
using Plots
import Random: randperm
using MAT
using Images
using BlockArrays
```

## Load Data

```julia
file = matopen("./face.mat")
Y = read(file, "Y")
close(file)
b = vec(Y);
```

```julia
Yplot = BlockArray{Float64}( undef, 192*ones(Int, 8), 168*ones(Int, 8))
for i in 1:64
    Yplot[Block(i)] .= Y[:,:,i]
end
colorview(Gray, Yplot)
```

## Construct atomic sets

```julia
m = 192 * 168
n = 64
ks = 210296
kr = 5
s = 6941.40
r = 796.86
As = OneBall(m*n; maxrank = ks); As = s*As
Ar = NucBall(m, n; maxrank = kr); Ar = r*Ar
A = As + Ar
```

## Solve demixing problem

```julia
opt = 265.61^2/2
dual_gaps = Vector{Float64}()
pr_infeas = Vector{Float64}()
sol = level_set(I(m*n), b, A; 
    =opt, tol=1e-3, maxIts=1000, max=1.0, dual_gaps=dual_gaps, infeas=pr_infeas);
```

```julia
x = constructPrimal(sol);
```

## Plot result

```julia
xr = reshape(x[2], 192, 168, 64)
xr ./= 0.5
Xplot = BlockArray{Float64}( undef, 192*ones(Int, 8), 168*ones(Int, 8))
for i in 1:64
    Xplot[Block(i)] .= xr[:,:,i]
end
colorview(Gray, Xplot)
```
